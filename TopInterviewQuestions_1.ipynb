```python
# 1. Two Sum
# Hash table, dictionary, key - value
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        index = {}
        for idx, val in enumerate(nums):
            if target - val not in index:
                index[val] = idx
            else:
                return [idx, index[target - val]]
```

```python
# 2. Add Two Numbers
# linked list, end condition includes carry = 0
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(-1)
        tail = dummy
        carry = 0
        
        while l1 or l2 or carry:
            if not l1:
                val1 = 0
            else:
                val1 = l1.val
            if not l2:
                val2 = 0
            else:
                val2 = l2.val
                
            num = val1 + val2 + carry
            if num < 10:
                new_node = ListNode(num)
                carry = 0
            else:
                num = num % 10
                carry = 1
                new_node = ListNode(num)
            
            tail.next = new_node
            tail = tail.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None
        
        return dummy.next
```

```python
# 3. Longest Substring Without Repeating Characters
# two pointers, the same direction, set, update (add), delete (remove)
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        chars = set()
        left = 0
        cnt = 0
        
        for right in range(len(s)):
            while s[right] in chars:
                chars.remove(s[left])
                left += 1
            chars.add(s[right])
            cnt = max(cnt, len(chars))
        
        return cnt
```

```python
# 8. String to Integer (atoi)
class Solution:
    def myAtoi(self, s: str) -> int:
        s = s.strip()
        if s == '': 
            return 0
        ls = list(s)
        #print(ls)
        
        if ls[0] == '-':
            sign = -1
        else:
            sign = 1
        
        if ls[0] == '-' or ls[0] == '+': 
            del ls[0]
            
        res = ''
        i = 0
        while i < len(ls) and ls[i].isdigit():
            res += ls[i]
            i += 1
        if len(res) == 0:
            res = 0
        else:
            res = sign * int(res)
        
        if res >= 2 ** 31 - 1:
            return 2 ** 31 - 1
        elif res <= -2 ** 31:
            return -2 ** 31
        else:
            return res
```

```python
# 11. Container With Most Water
# two pointers, from ends to middle
class Solution:
    def maxArea(self, height: List[int]) -> int:
        mx, i, j = 0, 0, len(height) - 1
        while i < j:
            if height[i] < height[j]:
                mx = max(mx, (j - i) * height[i])
                i += 1
            else:
                mx = max(mx, (j - i) * height[j])
                j -= 1
        return mx   
```

```python
# 13. Roman to Integer
# one pointer
class Solution:
    def romanToInt(self, S: str) -> int:
        ROMAN = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
        
        if S == "":
            return 0
            
        index = len(S) - 2
        sum = ROMAN[S[-1]]
        while index >= 0:
            if ROMAN[S[index]] < ROMAN[S[index + 1]]:
                sum -= ROMAN[S[index]]
            else:
                sum += ROMAN[S[index]]
            index -= 1
        return sum
```

```python
# 14. Longest Common Prefix
# string, CRUD
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if len(strs) <= 1:
            return strs[0] if len(strs) == 1 else ""
        end, minl = 0, min([len(s) for s in strs])
        while end < minl:
            for i in range(1, len(strs)):
                if strs[i][end] != strs[i-1][end]:
                    return strs[0][:end]
            end = end + 1
        return strs[0][:end]
```

```python
# 15. 3Sum
#### 1. two pointers, enumeration, confirm the first num, use the rest list to provide 0 - first, skip if nums[i] == nums[i - 1]
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums = sorted(nums)
        result = set()
        
        for i in range(len(nums)):
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            
            l = i + 1
            r = len(nums) - 1
            target = 0 - nums[i]
            while l < r:
                if nums[l] + nums[r] == target:
                    result.add((nums[i], nums[l], nums[r]))
                    l += 1
                    r -= 1
                elif nums[l] + nums[r] < target:
                    l += 1
                else:
                    r -= 1
        return list(result)
```

```python
# 17. Letter Combinations of a Phone Number
# recursion
KEYBOARD = {
    '2': 'abc',
    '3': 'def',
    '4': 'ghi',
    '5': 'jkl',
    '6': 'mno',
    '7': 'pqrs',
    '8': 'tuv',
    '9': 'wxyz'
}

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        
        res = []
        self.dfs(digits, 0, [], res)
        
        return res
    
    def dfs(self, digits, index, chars, res):
        if index == len(digits):
            res.append(''.join(chars))
            return
        
        for letter in KEYBOARD[digits[index]]:
            chars.append(letter)
            self.dfs(digits, index + 1, chars, res)
            chars.pop()
```

```python
# 19. Remove Nth Node From End of List
#### easy, dummy, slow, and fast
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        if not head:
            return None
        
        dummy = ListNode(-1) 
        dummy.next = head
        slow = dummy
        fast = dummy
        
        for _ in range(n):
            fast = fast.next
        
        while fast.next:
            slow = slow.next
            fast = fast.next
        
        slow.next = slow.next.next
        
        return dummy.next
```

```python
# 20. Valid Parentheses
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        dict = {"]":"[", "}":"{", ")":"("}
        for char in s:
            if char in dict.values():
                stack.append(char)
            elif char in dict.keys():
                if stack == [] or dict[char] != stack.pop():
                    return False
            else:
                return False
        return stack == []
```

```python
# 21. Merge Two Sorted Lists
class Solution:
    def mergeTwoLists(self, list1, list2):
        dummy = ListNode(0)
        tmp = dummy
        while list1 != None and list2 != None:
            if list1.val < list2.val:
                tmp.next = list1
                list1 = list1.next
            else:
                tmp.next = list2
                list2 = list2.next
            tmp = tmp.next
        if list1 != None:
            tmp.next = list1
        else:
            tmp.next = list2
        return dummy.next
```

```python
# 22. Generate Parentheses
#### draw a decision tree when n == 2, return and run the last row then return the former recursion
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        if n == 0:
            return []
        
        res = []
        self.helpler(n, n, '', res)
        
        return res

    def helpler(self, l, r, item, res):
        if r < l:
            return
        
        if l == 0 and r == 0:
            res.append(item)
            
        if l > 0:
            self.helpler(l - 1, r, item + '(', res)
            
        if r > 0:
            self.helpler(l, r - 1, item + ')', res)
```

```python
# 28. Implement strStr()
#### string.find(substring) return substring's index
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if haystack is None or needle is None:
            return None
        
        return haystack.find(needle)
```
