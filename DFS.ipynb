```python
# 938. Range Sum of BST
class Solution:
    def rangeSumBST(self, root: Optional[TreeNode], low: int, high: int) -> int:
        self.res = 0
        self.low = low
        self.high = high
        self.helper(root)
        return self.res
        
    def helper(self, root):
        if root is None:
            return None
        if root.val >= self.low and root.val <= self.high:
            self.res += root.val
        
        self.helper(root.left)
        self.helper(root.right)
```

```python
# 617. Merge Two Binary Trees
class Solution:
    def mergeTrees(self, root1: Optional[TreeNode], root2: Optional[TreeNode]) -> Optional[TreeNode]:
        if root1 is None and root2 is None :
            return None
        
        if root1 is None and root2 is not None:
            return root2
        
        if root2 is None and root1 is not None:
            return root1
        
        new_node = TreeNode()
        new_node.val = root1.val + root2.val
        new_node.left = self.mergeTrees(root1.left, root2.left)
        new_node.right = self.mergeTrees(root1.right, root2.right)
        return new_node
```

```python
# 897. Increasing Order Search Tree
class Solution:
    def increasingBST(self, root: TreeNode) -> TreeNode:
        result = self.bst = TreeNode(-1, None, None)
        self.helper(root)
        
        return result.right
    
    def helper(self, node):
    
        if node is None:
            return None
        
        self.helper(node.left)
        
        new_node = TreeNode(node.val, None, None)
        self.bst.right = new_node
        self.bst = self.bst.right
        
        self.helper(node.right)
```

```python
# 563. Binary Tree Tilt
class Solution:
    def findTilt(self, root):
        self.res = 0
        self.helper(root)
        return self.res
    
    def helper(self, root):
        if not root:
            return 0
        left = self.helper(root.left)
        right = self.helper(root.right)
        self.res += abs(left - right)
        
        return left + right + root.val
```

```python
# 257. Binary Tree Paths
class Solution:
    def binaryTreePaths(self, root: Optional[TreeNode]) -> List[str]:
        self.res = []
        curr = ''
        self.helper(root, curr)
        return self.res
    
    def helper(self, root, s):
        if not root:
            return None
        
        if not root.left and not root.right:
            if len(s) == 0:
                s += str(root.val)
            else:
                s += ',' + str(root.val)
                s = s.replace(',', '->')
            self.res.append(s)
        
        if len(s) == 0:
            s += str(root.val)
        else:
            s += ',' + str(root.val)
        
        self.helper(root.left, s)
        self.helper(root.right, s)
```

```python
# 606. Construct String from Binary Tree
class Solution:
    def tree2str(self, root):
        return self.helper(root)
    
    def helper(self, root):
        if not root:
            return None
        
        l = ''
        r = ''
        
        if root.left:
            l = '(' + self.helper(root.left) + ')'
            
        if root.right:
            if l == '':
                r = '()'
            r += '(' + self.helper(root.right) + ')'
        
        return str(root.val) + l + r
```

```python
# 100. Same Tree
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        return self.helper(p, q)
        
    def helper(self, p, q):
        if p is None and q is None:
            return True
        if p is None and not q is None:
            return False
        if p is not None and q is None:
            return False
        
        if p.val != q.val:
            return False
        else:
            return self.helper(p.left, q.left) and self.helper(p.right, q.right)
```

```python
# 543. Diameter of Binary Tree
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        self.max_depth = 0
        self.helper(root)
        
        return self.max_depth - 1
    
    def helper(self, root):
        if root is None:
            return 0
        
        left = self.helper(root.left)
        right = self.helper(root.right)
        
        self.max_depth = max(self.max_depth, left + right + 1)
        
        return max(left, right) + 1
```

```python
# 993. Cousins in Binary Tree
class Solution:
    def isCousins(self, root: Optional[TreeNode], x: int, y: int) -> bool:
        self.x = x
        self.y = y
        self.x_depth = 0
        self.y_depth = 0
        self.x_parent = 0
        self.y_parent = 0
        self.helper(root, 1, root.val)
        if self.x_depth == self.y_depth and self.x_parent != self.y_parent:
            return True
        else:
            return False
    
    def helper(self, root, depth, parent):
        if root is None:
            return None
        
        if root.val == self.x:
            self.x_depth = depth
            self.x_parent = parent
            
        if root.val == self.y:
            self.y_depth = depth
            self.y_parent = parent
        
        
        self.helper(root.left, depth + 1, root.val)
        self.helper(root.right, depth + 1, root.val)
```

```python
# 110. Balanced Binary Tree
class Solution:
    def isBalanced(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True

        if not self.isBalanced(root.left):
            return False
        if not self.isBalanced(root.right):
            return False
    
        return abs(self.get_height(root.left) - self.get_height(root.right)) <= 1

    def get_height(self, root):
        if root is None:
            return 0
    
        return max(self.get_height(root.left), self.get_height(root.right)) + 1
```
